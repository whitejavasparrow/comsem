```python
sentence_lst = article.split("\n")
sentence_lst
```

```python
def tagger(text):

    ws_results = ws([text])
    pos_results = pos([text])
    ner_results = ner([text], pos_results)

    return {"ws_results": ws_results,
            "pos_results": pos_results,
            "ner_results": ner_results}
```

```python
sentence_lst_seg = []
for sentence in sentence_lst:
    sentence_lst_seg.append(tagger(sentence))
    
sentence_lst_seg
```

```python
date_lst, time_lst = [], []

for sentence in sentence_lst_seg:
    for entities in sentence["ner_results"]:
        for entity in entities:
            if entity[2] == "DATE":
                date_lst.append(entity[3])
            elif entity[2] == "TIME":
                time_lst.append(entity[3])
                
print(date_lst)
print(time_lst)
```
